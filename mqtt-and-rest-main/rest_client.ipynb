```
import requests
import requests
import matplotlib.pyplot as plt

host = 'https://1971cb16-e709-4b3f-8d6d-f6f260f3f6b7.deepnoteproject.com'
# host = 'http://localhost:8080'
try:
    response = requests.get(host + '/status')
except Exception as e:
    print("Connection not built")
    exit()
    
if response.status_code == 200:
    status = response.json()['status']
    print(f'The server is {status}.')
else:
    print('The server is offline.')
    exit()

payload = {         
    'from':'1674680023223',
    'to': '1674680033255'
}
response = requests.get(host + '/write_resource')
devices_list = []
if response.status_code == 200:
    print("Connected devices are:")
    devices_list = response.json()['mac_addresses']
    print(devices_list)

response = requests.get(host + '/device/0x51888af87868', params=payload)
if response.status_code == 200:
    item = response.json()
    mac = item['mac_address']   
    timeStamp = item['timestamps']
    battery_level = item['battery_levels']  
    pluged = item['power_plugged']

    print(f'MAC address: {mac}')
    print(f'TimeStamp: {timeStamp}')
    print(f'Battery Level: {battery_level}')
    print(f'Power Pluged: {pluged}')

    plt.plot(timeStamp, battery_level)
    # plt.ylim(75.9,76.1)
    plt.title('battery level')
    plt.xlabel('time stamp')
    plt.ylabel('level')
    plt.show()
    # plt.savefig('my_plot.png')
if len(devices_list)> 0:
    last_device = devices_list[-1]
    response = requests.delete(host + f'/device/{last_device}')

# response = requests.get(host + '/write_resource')
# if response.status_code == 200:
#     item = response.json()

#     print(item)
    
```
